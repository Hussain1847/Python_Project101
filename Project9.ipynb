```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the sample CSV file
file_path = "/content/sample_data.csv"
df = pd.read_csv(file_path)

# Show data preview
print("\nData Preview:")
print(df.head())

# Show data summary
print("\nData Summary:")
print(df.describe())

# Select column for filtering
columns = df.columns.tolist()
print("\nAvailable columns:", columns)

selected_column = input("Enter column name to filter by: ")
if selected_column in columns:
    unique_values = df[selected_column].unique()
    print(f"Unique values in {selected_column}: {unique_values}")

    selected_value = input(f"Enter a value from {selected_column} to filter: ")
    try:
        filtered_df = df[df[selected_column] == selected_value]
    except:
        filtered_df = df[df[selected_column] == float(selected_value)]

    print("\nFiltered Data:")
    print(filtered_df.head())
else:
    print("Invalid column selected.")

# Plot data
x_column = input("Enter column for x-axis: ")
y_column = input("Enter column for y-axis: ")

if x_column in columns and y_column in columns:
    plt.figure(figsize=(10, 5))
    plt.plot(filtered_df[x_column], filtered_df[y_column], marker="o", linestyle="-", color="b")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title(f"{y_column} vs {x_column}")
    plt.grid()
    plt.show()
else:
    print("Invalid columns selected for plotting.")


```
